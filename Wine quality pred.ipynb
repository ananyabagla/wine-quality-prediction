
**Decision Tree chooses the root node by utilizing two methods- gini and entropy-**
*   Calculate gini for each class
*   gini = summation_of(probability)square

*   gini is gini impurity
*   class with least gini will be least impure and will be selected

**Entropy and Information Gain**
Column with the most IG or least Entropy is selected as the root node

**CART**(Classificstion and Regression Tree) is an algo used within Decision tress for both classification and regression purposes.

**iv3** is an algo of decision tree used only for regression

**Overfitting** - When the score of our model with x_train and y_train is like 90.0 but with x_test and y_test (unknown data) , comes out to be lower (like 85.0) , that situation is called as over-fitting

In decision tree we can fix the problem of over-fitting by **Pruning**

**Ensemble Tech** - is an approach where we use multiple instances of the same algo or a combination of multiple algos to make the decision. We ask multiple decision makers the decision (for ex -in the case of voting , to avoid bias)

Ensemble tech is further divided into -

*  **Bagging Tech** aka Bootstrapping(coz it follows boot strapping) - voting, multiple decision maker , everyone has same priority and decision with the max votes wins.

---
We can have any algo inside the bagging its just that we can only use 1 type of certain algo so if we have a bag with svm base model then it can have 100 diff svm models

---



*   **Boosting Tech** - multiple decision maker but we try to find the better decion maker and his decision gets more weight.(student(0.001) vs teacher(15) vs industry speacilist(84.999)). In Boosting we have a weak classifier(classifier whoose decision is slightly better than a random guess) , it passes its decision to a batter classifer and that classifier to a even better classifier and it keeps on going thats why it is called boosting

*Stacking* is another approach , not very popular - similar to bagging but here the nature of decision maker is heterogeneous(mul diff algos - LR , DT , SVM , KNN etc used in esemble tech).

2 approaches to divide data_set:

*   ***Bootstrapping*** - here there is a high chance of repetition of some data in the smaller didided data_sets
*   ***Pasting*** - No repetition , no overlap



**Random Forest Algo** is a part of Ensemble Technique Bagging .
We can use any algo(knn,svm,LR,N Bayes,etc) inside the bagging technique but normally Random forest is used.


---
Random Forest is a combination of multiple decision trees


---
**IN BOOSTING -(ADA BOOSTING/aka ADAPTIVE BOOSTING)**

*   Weak classifier gives the same weight to each and every data set
*   Then we calculate error value and store it as alpha for where we have an error(y_hat - y)
*   Then we update the value of weights where-ever we have an error
*   We have to control the weight expansion such that after normalizing the new weight the summation of all the weights remains equal to 1
*   Now we send this data to our new model , then we repeat the same process -> get new weights , update data then send it to the next model.

* At the end compare the summation of all the results from all the trees and the final result is either the one with highest sum(for regression), or the one with most weighted voted avg(for classification)



---
**Boosting is of 3 types-**

* Ada Boost
* Gtadient Boost
* XG(extreme gradient) Boost (better ver of Grad Boost)

# Removing over-fitting - Pruning

Manual Method-

Lets perform post prunning

```python
path = Dt.cost_complexity_pruning_path(x_train,y_train)
```

```python
ccp_alpha = path['ccp_alphas']
```

```python
ccp_alpha
```

```python
path
```

```python
Dt_model2 = []
for ccp in ccp_alpha:
  dt_m = DecisionTreeClassifier(ccp_alpha = ccp)
  dt_m.fit(x_train , y_train)
  Dt_model2.append(dt_m)
```

```python
Dt_model2
```

```python
model_score_test = []
model_score_train = []
for i in Dt_model2:
  model_score_test.append([ i , i.score(x_test , y_test) ])
  model_score_train.append([ i , i.score(x_train , y_train) ])


```

```python
model_score_train
```

```python
model_score_test
```

```python
test_score = []
train_score = []
for i in Dt_model2:
  test_score.append( i.score(x_test , y_test) )
  train_score.append( i.score(x_train , y_train))
```

```python
test_score
```

```python
train_score
```

Now we have to select the ccp_alpha values where test_score and train_score are close to each otherto

```python
fig,ax = plt.subplots()
ax.set_xlabel("aplha")
ax.set_ylabel("accuracy")
ax.plot(ccp_alpha , train_score , 'r-+'  , label = 'train' , drawstyle = "steps-post")
ax.plot(ccp_alpha , test_score , 'b-' , label = 'validation' , drawstyle = "steps-post")
ax.legend()
plt.show()
```

```python
dt_model_ccp = DecisionTreeClassifier(random_state=0 , ccp_alpha = 0.024)
dt_model_ccp.fit(x_train , y_train)
```

```python
dt_model_ccp.score(x_train , y_train)
```

```python
dt_model_ccp.score(x_test , y_test)
```
